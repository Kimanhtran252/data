```python
data = pd.read_csv('774b823f-ad32-4372-964b-fd8652b20d21_Data.csv')
data = data.rename(columns={'2011 [YR2011]':'2011','2012 [YR2012]':'2012', '2013 [YR2013]':'2013','2014 [YR2014]':'2014','2015 [YR2015]':'2015','2016 [YR2016]':'2016','2017 [YR2017]':'2017',	'2018 [YR2018]':'2018','2019 [YR2019]':'2019',	'2020 [YR2020]':'2020'})
data = data.drop(columns = ['Country Code','Series Code'])
```

```python
data.head(5)
```

```python
data = data.set_index('Series Name')
```

```python
po =  data.loc[['Population, total']]
po = po.reset_index('Series Name')
po1 = po.melt(id_vars=['Country Name','Series Name'])
```

```python
fig1= px.bar(po1,x='variable',y='value', color='Country Name', barmode='group', labels={'variable':'Year','value':'Population'},
             category_orders={"Country Name":["Indonesia","Vietnam","Thailand"]})
fig1.update_layout(bargap=0.3, title='Population total from 2010-2019', yaxis_title_text='Population', hoverlabel = dict(bgcolor='white',font_size=14),
 showlegend= True,
 title_font_size = 25,
 title_font_color='black',
 title_x =0.4,
 xaxis= dict(tickfont_size =14),
 yaxis = dict(tickfont_size=14))
fig1.update_layout(
    legend=dict(
        traceorder="reversed",
        bgcolor="white",
        bordercolor="Black",
        borderwidth=2
    )
)
fig1.update_traces(marker_line_color='rgb(8,48,107)',
                  marker_line_width=1.5, opacity=0.6)
 
fig1.show()
```
